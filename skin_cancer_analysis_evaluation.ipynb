# Loading Best Models

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Metrics
from sklearn.metrics import accuracy_score, mean_squared_error, f1_score, classification_report

#ROC-AUC score + ROC Curve
from sklearn.metrics import RocCurveDisplay
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report

# Confusion Matrix
from sklearn.metrics import confusion_matrix

# Displaying Confusion Matrices
from sklearn.metrics import ConfusionMatrixDisplay

# Load Logistic Regression (OvR EN)
with open("ovr_en_logreg.model", "rb") as file:
    best_logreg = pickle.load(file)

# Load Hist Gradient Boosting
with open("hist_gradient_boosting.model", "rb") as file:
    best_tree = pickle.load(file)

# Load Neural Network
with open("Best_Neural_Network.model", "rb") as file:
    best_nn = pickle.load(file)
```

# LR OvR EN: Accuracy, precision, recall, f1-score

```python
X_train, X_test = pickle.load(open("skin_cancer_features.data", "rb"))
y_train, y_test = pickle.load(open("skin_cancer_labels.data", "rb"))

just_in_case = y_test

# Sanity Check
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test (current) shape:", y_test.shape)
print("original y_test (just_in_case) shape:", just_in_case.shape)

with open("ovr_en_logreg.model", "rb") as file:
    best_logreg = pickle.load(file)

y_pred_ovr_en = best_logreg.predict(X_test)

# Accuracy
accuracy_ovr_en = accuracy_score(just_in_case, y_pred_ovr_en)
print("Accuracy for OvR EN Logistic Regression:", accuracy_ovr_en)
print("Classification report for OvR EN Logistic Regression:")
print(classification_report(just_in_case, y_pred_ovr_en))
```

# Hist Gradient Boosting: Accuracy, precision, recall, f1-score

```python
# Sanity Check

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test (current) shape:", y_test.shape)
print("original y_test (just_in_case) shape:", just_in_case.shape)

with open ("hist_gradient_boosting.model", 'rb') as file:
    best_tree = pickle.load(file)

y_predicted_hist_gradient_boosting_model = best_tree.predict(X_test)

# Accuracy
accuracy_hist = accuracy_score (just_in_case, y_predicted_hist_gradient_boosting_model)
print("Accuracy for hist_gradient_boostin:", accuracy_hist)
print("Printing classification report for hist_gradient_boosting:")
print(classification_report(just_in_case, y_predicted_hist_gradient_boosting_model))
```

# Best Neural Network: Accuracy, precision, recall, f1-score

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test (current) shape:", y_test.shape)
print("original y_test (just_in_case) shape:", just_in_case.shape)

with open ("Best_Neural_Network.model", 'rb') as file:
  best_nn_model = pickle.load(file)

best_nn_model_y_pred = best_nn_model.predict(X_test)

predicted_classes_nn = np.argmax(best_nn_model_y_pred, axis=1)

best_model_accuracy = accuracy_score (just_in_case, predicted_classes_nn)

print("Accuracy for Neural Network Model:", best_model_accuracy)
print("Classification report for Neural Network Model:")
print(classification_report(just_in_case, predicted_classes_nn))
```

# ROC Curve + AUC Score

```python
y_test = just_in_case
classes = np.unique(y_test)
print("Classes:", classes)
n_classes = len(classes)
```

```python
y_test_bin = label_binarize(y_test, classes=classes)
print("y_test_bin shape:", y_test_bin.shape)
```

```python
# OvR EN Logistic Regression
y_prob_logreg = best_logreg.predict_proba(X_test)

# Hist Gradient Boosting
y_prob_tree = best_tree.predict_proba(X_test)

# Neural Network: already outputs probs
y_prob_nn = best_nn.predict(X_test)
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

for i in range(n_classes):
    RocCurveDisplay.from_predictions(
        y_true = y_test_bin[:, i],
        y_pred = y_prob_logreg[:, i],
        name = f"Class {classes[i]} vs rest",
        ax = ax
    )

ax.plot([0, 1], [0, 1], "k--", label="Chance level")
ax.set_title("Multiclass ROC – OvR EN Logistic Regression")
ax.legend(loc="lower right")
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

for i in range(n_classes):
    RocCurveDisplay.from_predictions(
        y_true = y_test_bin[:, i],
        y_pred = y_prob_tree[:, i],
        name = f"Class {classes[i]} vs rest",
        ax = ax
    )

ax.plot([0, 1], [0, 1], "k--", label="Chance level")
ax.set_title("Multiclass ROC – Hist Gradient Boosting")
ax.legend(loc="lower right")
ax.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

for i in range(n_classes):
    RocCurveDisplay.from_predictions(
        y_true = y_test_bin[:, i],
        y_pred = y_prob_nn[:, i],
        name = f"Class {classes[i]} vs rest",
        ax = ax
    )

ax.plot([0, 1], [0, 1], "k--", label="Chance level")
ax.set_title("Multiclass ROC – Neural Network")
ax.legend(loc="lower right")
ax.grid(True)
plt.show()
```

ROC-AUC

```python
# For Logistic Regression OvR
y_prob_logreg = best_logreg.predict_proba(X_test)
roc_auc_logreg = roc_auc_score(y_test_bin, y_prob_logreg, multi_class='ovr')
print("Logistic Regression ROC-AUC:", roc_auc_logreg)

# For Decision Tree
y_prob_tree = best_tree.predict_proba(X_test)
roc_auc_tree = roc_auc_score(y_test_bin, y_prob_tree, multi_class='ovr')
print("Decision Tree ROC-AUC:", roc_auc_tree)

# For Neural Network
y_prob_nn = best_nn.predict(X_test)
roc_auc_nn = roc_auc_score(y_test_bin, y_prob_nn, multi_class='ovr')
print("Neural Network ROC-AUC:", roc_auc_nn)
```

# Confusion Matrices

```python
# Confusion Matrix
y_pred_lr = best_logreg.predict(X_test)
cm_lr = confusion_matrix(y_test, y_pred_lr)

print("Confusion Matrix (Logistic Regression):")
print(cm_lr)

# NN model 2
y_pred_nn = np.argmax(best_nn.predict(X_test), axis=1)
cm_nn = confusion_matrix(y_test, y_pred_nn)

print("Confusion Matrix (Neural Network):")
print(cm_nn)

# Hist gradient
y_pred_tree = best_tree.predict(X_test)
cm_tree = confusion_matrix(y_test, y_pred_tree)

print("Confusion Matrix (Hist Gradient Boosting):")
print(cm_tree)
```

```python
# Logistic Regression predictions
y_pred_lr = best_logreg.predict(X_test)

# Confusion matrix
cm_lr = confusion_matrix(y_test, y_pred_lr)

# Plot
disp = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=classes)
disp.plot(cmap=plt.cm.Blues)

plt.title("Confusion Matrix – Logistic Regression (OvR EN)")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Neural Network predictions (probabilities -> class labels)
y_prob_nn = best_nn.predict(X_test)
y_pred_nn = np.argmax(y_prob_nn, axis=1)

# Confusion matrix
cm_nn = confusion_matrix(y_test, y_pred_nn)

# Plot
disp_nn = ConfusionMatrixDisplay(confusion_matrix=cm_nn, display_labels=classes)
disp_nn.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix – Neural Network")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Hist Gradient Boosting predictions
y_pred_tree = best_tree.predict(X_test)

# Confusion matrix
cm_tree = confusion_matrix(y_test, y_pred_tree)

# Plot
disp_tree = ConfusionMatrixDisplay(confusion_matrix=cm_tree, display_labels=classes)
disp_tree.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix – Hist Gradient Boosting")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```
